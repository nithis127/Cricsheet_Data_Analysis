```python
import pymysql
class cricsheet_db:    
    #---constructor---
    def __init__(self,host,user,password,database,port=3306,autocommit=True): 
        connection=pymysql.connect(host=host,
                      user=user,
                      password=password,
                      port=port)
        cursor=connection.cursor()
        cursor.execute(f'CREATE DATABASE IF NOT EXISTS {database}')     #create database if not exists
        cursor.close()
        connection.close()

        self.connection=pymysql.connect(                 #conneting Python to MySQL database
            host=host,
            user=user,
            password=password,
            database=database,
            port=port,
            autocommit=autocommit
            )
        self.cursor=self.connection.cursor()

    #---create IPL matches table if not exists---
    def create_ipl_matches_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS IPL_MATCHES (
                    MATCH_ID VARCHAR(20),
                    START_DATE DATE,
                    END_DATE DATE,
                    SEASON VARCHAR(15),
                    MATCH_NUMBER INT,
                    MATCH_STAGE VARCHAR(20),
                    MATCH_NAME VARCHAR(100),
                    TEAMS VARCHAR(250),
                    MATCH_WINNER VARCHAR(150),
                    WIN_BY_WICKETS INT,
                    WIN_BY_RUNS INT,
                    ELIMINATOR VARCHAR(150),
                    RESULT VARCHAR(50),
                    PLAYER_OF_MATCH VARCHAR(50),
                    VENUE VARCHAR(200),
                    CITY VARCHAR(50),
                    TOSS_WIN VARCHAR(150),
                    TOSS_DECISION VARCHAR(10),
                    CONSTRAINT PK_IPL PRIMARY KEY (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('IPL Matches Table created')

    #---create ODI matches table if not exists---
    def create_odi_matches_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS ODI_MATCHES (
                    MATCH_ID VARCHAR(20),
                    START_DATE DATE,
                    END_DATE DATE,
                    SEASON VARCHAR(15),
                    MATCH_NUMBER INT,
                    MATCH_STAGE VARCHAR(20),
                    MATCH_NAME VARCHAR(100),
                    TEAMS VARCHAR(250),
                    MATCH_WINNER VARCHAR(150),
                    WIN_BY_WICKETS INT,
                    WIN_BY_RUNS INT,
                    ELIMINATOR VARCHAR(150),
                    RESULT VARCHAR(50),
                    PLAYER_OF_MATCH VARCHAR(50),
                    VENUE VARCHAR(200),
                    CITY VARCHAR(50),
                    TOSS_WIN VARCHAR(150),
                    TOSS_DECISION VARCHAR(10),
                    CONSTRAINT PK_ODI PRIMARY KEY (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('ODI Matches Table created')

    #---create T20 matches table if not exists---
    def create_t20_matches_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS T20_MATCHES (
                    MATCH_ID VARCHAR(20),
                    START_DATE DATE,
                    END_DATE DATE,
                    SEASON VARCHAR(15),
                    MATCH_NUMBER INT,
                    MATCH_STAGE VARCHAR(20),
                    MATCH_NAME VARCHAR(100),
                    TEAMS VARCHAR(250),
                    MATCH_WINNER VARCHAR(150),
                    WIN_BY_WICKETS INT,
                    WIN_BY_RUNS INT,
                    ELIMINATOR VARCHAR(150),
                    RESULT VARCHAR(50),
                    PLAYER_OF_MATCH VARCHAR(50),
                    VENUE VARCHAR(200),
                    CITY VARCHAR(50),
                    TOSS_WIN VARCHAR(150),
                    TOSS_DECISION VARCHAR(10),
                    CONSTRAINT PK_T20 PRIMARY KEY (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('T20 Matches Table created')

    #---create T20 matches table if not exists---
    def create_test_matches_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS TEST_MATCHES (
                    MATCH_ID VARCHAR(20),
                    START_DATE DATE,
                    END_DATE DATE,
                    SEASON VARCHAR(15),
                    MATCH_NUMBER INT,
                    MATCH_STAGE VARCHAR(20),
                    MATCH_NAME VARCHAR(100),
                    TEAMS VARCHAR(250),
                    MATCH_WINNER VARCHAR(150),
                    WIN_BY_WICKETS INT,
                    WIN_BY_RUNS INT,
                    WIN_BY_INNINGS INT,
                    RESULT VARCHAR(50),
                    PLAYER_OF_MATCH VARCHAR(50),
                    VENUE VARCHAR(200),
                    CITY VARCHAR(50),
                    TOSS_WIN VARCHAR(150),
                    TOSS_DECISION VARCHAR(10),
                    CONSTRAINT PK_TEST PRIMARY KEY (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('TEST Matches Table created')

    #---create IPL deliveries table if not exists---
    def create_ipl_deliveries_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS IPL_DELIVERIES (
                    MATCH_ID VARCHAR(20),
                    INNING INT,
                    TEAM VARCHAR(150),
                    OVER_NUMBER INT,
                    BALL_NUMBER FLOAT,
                    BATTER VARCHAR(50),
                    BOWLLING_TEAM VARCHAR(150),
                    BOWLER VARCHAR(50),
                    NON_STRIKER VARCHAR(50),
                    EXTRAS VARCHAR(20),
                    BATTER_RUNS INT,
                    RUNS INT,
                    WICKET VARCHAR(50),
                    PLAYER_OUT VARCHAR(50),
                    FIELDER_1 VARCHAR(50),
                    FIELDER_2 VARCHAR(50),
                    FIELDER_3 VARCHAR(50),
                    CONSTRAINT FK_IPL FOREIGN KEY (MATCH_ID) REFERENCES IPL_MATCHES (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('IPL Deliveries Table created')

    #---create ODI deliveries table if not exists---
    def create_odi_deliveries_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS ODI_DELIVERIES (
                    MATCH_ID VARCHAR(20),
                    INNING INT,
                    TEAM VARCHAR(150),
                    OVER_NUMBER INT,
                    BALL_NUMBER FLOAT,
                    BATTER VARCHAR(50),
                    BOWLLING_TEAM VARCHAR(150),
                    BOWLER VARCHAR(50),
                    NON_STRIKER VARCHAR(50),
                    EXTRAS VARCHAR(20),
                    BATTER_RUNS INT,
                    RUNS INT,
                    WICKET VARCHAR(50),
                    PLAYER_OUT VARCHAR(50),
                    FIELDER_1 VARCHAR(50),
                    FIELDER_2 VARCHAR(50),
                    FIELDER_3 VARCHAR(50),
                    FIELDER_4 VARCHAR(50),
                    CONSTRAINT FK_ODI FOREIGN KEY (MATCH_ID) REFERENCES ODI_MATCHES (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('ODI Deliveries Table created')

    #---create T20 deliveries table if not exists---
    def create_t20_deliveries_table(self):                          
        query = '''CREATE TABLE IF NOT EXISTS T20_DELIVERIES (
                    MATCH_ID VARCHAR(20),
                    INNING INT,
                    TEAM VARCHAR(150),
                    OVER_NUMBER INT,
                    BALL_NUMBER FLOAT,
                    BATTER VARCHAR(50),
                    BOWLLING_TEAM VARCHAR(150),
                    BOWLER VARCHAR(50),
                    NON_STRIKER VARCHAR(50),
                    EXTRAS VARCHAR(20),
                    BATTER_RUNS INT,
                    RUNS INT,
                    WICKET VARCHAR(50),
                    PLAYER_OUT VARCHAR(50),
                    FIELDER_1 VARCHAR(50),
                    FIELDER_2 VARCHAR(50),
                    FIELDER_3 VARCHAR(50),
                    CONSTRAINT FK_T20 FOREIGN KEY (MATCH_ID) REFERENCES T20_MATCHES (MATCH_ID));
                '''  
        self.cursor.execute(query)
        print('T20 Deliveries Table created')

    #---create TEST deliveries table if not exists---
    def create_test_deliveries_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS TEST_DELIVERIES (
                    MATCH_ID VARCHAR(20),
                    INNING INT,
                    TEAM VARCHAR(150),
                    OVER_NUMBER INT,
                    BALL_NUMBER FLOAT,
                    BATTER VARCHAR(50),
                    BOWLLING_TEAM VARCHAR(150),
                    BOWLER VARCHAR(50),
                    NON_STRIKER VARCHAR(50),
                    EXTRAS VARCHAR(20),
                    BATTER_RUNS INT,
                    RUNS INT,
                    WICKET VARCHAR(50),
                    PLAYER_OUT VARCHAR(50),
                    FIELDER_1 VARCHAR(50),
                    FIELDER_2 VARCHAR(50),
                    FIELDER_3 VARCHAR(50),
                    CONSTRAINT FK_TEST FOREIGN KEY (MATCH_ID) REFERENCES TEST_MATCHES (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('TEST Deliveries Table created')

#---create IPL deliveries table if not exists---
    def create_ipl_players_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS IPL_PLAYERS (
                    MATCH_ID VARCHAR(20),
                    TEAM VARCHAR(150),
                    PLAYER VARCHAR(150),
                    CONSTRAINT FK_IPL_PLAYER FOREIGN KEY (MATCH_ID) REFERENCES IPL_MATCHES (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('IPL Players Table created')

#---create ODI deliveries table if not exists---
    def create_odi_players_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS ODI_PLAYERS (
                    MATCH_ID VARCHAR(20),
                    TEAM VARCHAR(150),
                    PLAYER VARCHAR(150),
                    CONSTRAINT FK_ODI_PLAYER FOREIGN KEY (MATCH_ID) REFERENCES ODI_MATCHES (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('ODI Players Table created')

#---create T20 deliveries table if not exists---
    def create_t20_players_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS T20_PLAYERS (
                    MATCH_ID VARCHAR(20),
                    TEAM VARCHAR(150),
                    PLAYER VARCHAR(150),
                    CONSTRAINT FK_T20_PLAYER FOREIGN KEY (MATCH_ID) REFERENCES T20_MATCHES (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('T20 Players Table created')

#---create TEST deliveries table if not exists---
    def create_test_players_table(self):                          
        query =  '''CREATE TABLE IF NOT EXISTS TEST_PLAYERS (
                    MATCH_ID VARCHAR(20),
                    TEAM VARCHAR(150),
                    PLAYER VARCHAR(150),
                    CONSTRAINT FK_TEST_PLAYER FOREIGN KEY (MATCH_ID) REFERENCES TEST_MATCHES (MATCH_ID));
                 '''  
        self.cursor.execute(query)
        print('TEST Players Table created')
```
